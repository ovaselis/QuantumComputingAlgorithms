# Quantum Fourier Transform

**Quantum Fourier Transform** (QFT) is the quantum analogue of the discrete Fourier transform (DFT). It is a unitary (and therefore reversible) transformation that maps a quantum state from the numerical/index basis to the “frequency” basis. In practice, QFT is very important in algorithms that use interference and phase information (for example, phase estimation and Shor’s algorithm).

### Formulas

Suppose that we have an $N=2^n$ dimensional quantum state vector
$
{x}=(x_0,x_1,\dots,x_{N-1})^T
$
that represents the state
$$
|\psi\rangle=\sum_{j=0}^{N-1} x_j |j\rangle.
$$

The QFT of the state $|\psi\rangle$ is defined as
$$
|\phi\rangle=\sum_{k=0}^{N-1} y_k |k\rangle,
$$
where $y_k$ is defined as above
$$
y_k=\frac{1}{\sqrt{N}}\sum_{j=0}^{N-1} e^{\frac{2\pi i}{N}jk}\,x_j.
$$

Thus, for the basis state $|j\rangle$, which corresponds to the vector with $x_j=1$ and all other $x=0$, the QFT is
$$
\mathrm{QFT}_N|j\rangle=\frac{1}{\sqrt{N}}\sum_{k=0}^{N-1} e^{\frac{2\pi i}{N}jk}\,|k\rangle.
$$

$e^{\frac{2\pi i}{N}}$ is also often denoted by $\omega,$ and then the same expression can be written as
$$
\mathrm{QFT}_N|j\rangle=\frac{1}{\sqrt{N}}\sum_{k=0}^{N-1}\omega^{jk}\,|k\rangle.
$$

## Example: QFT on the basis state $|11\rangle$

If we have $n=2$ qubits, then $N=4$ and $\omega=e^{\frac{2\pi i}{4}}=i.$
The basis state $|11\rangle$ corresponds to the number $j=3$.
$$
\mathrm{QFT}_4|j\rangle=\frac{1}{\sqrt{4}}\sum_{k=0}^{3}\omega^{jk}|k\rangle
=\frac{1}{2}\sum_{k=0}^{3}\omega^{3k}|k\rangle.
$$

Let us calculate the coefficients for each $k$:
- $k=0$: $\omega^{3\cdot 0}=\omega^0=1$
- $k=1$: $\omega^{3\cdot 1}=\omega^3=i^3=-i$
- $k=2$: $\omega^{3\cdot 2}=\omega^6=i^6=-1$
- $k=3$: $\omega^{3\cdot 3}=\omega^9=i^9=i$

Thus:
$$
\mathrm{QFT}_4|11\rangle
=\frac{1}{2}\Big(|00\rangle - i|01\rangle - |10\rangle + i|11\rangle\Big).
$$

## Example: QFT on the state $\;\dfrac{|01\rangle+|11\rangle}{\sqrt2}$

Assume $n=2$ qubits, so $N=4$ and $\omega=e^{\frac{2\pi i}{4}}=i.$

The initial state is $|\psi\rangle=\frac{|01\rangle+|11\rangle}{\sqrt2}.$

Since the QFT is linear,
$$
\mathrm{QFT}_4|\psi\rangle
=\frac{1}{\sqrt2}\left(\mathrm{QFT}_4|01\rangle+\mathrm{QFT}_4|11\rangle\right).
$$

$$
\mathrm{QFT}_4|01\rangle
=\frac12\Big(|00\rangle+i|01\rangle-|10\rangle-i|11\rangle\Big),
$$
$$
\mathrm{QFT}_4|11\rangle
=\frac12\Big(|00\rangle-i|01\rangle-|10\rangle+i|11\rangle\Big).
$$

Let us add the terms by basis states:
- at $|00\rangle$: $\frac12+\frac12=1$
- at $|01\rangle$: $\frac{i}{2}+\left(-\frac{i}{2}\right)=0$
- at $|10\rangle$: $-\frac12+(-\frac12)=-1$
- at $|11\rangle$: $-\frac{i}{2}+\frac{i}{2}=0$

Therefore,
$$
\mathrm{QFT}_4|01\rangle+\mathrm{QFT}_4|11\rangle
=|00\rangle-|10\rangle.
$$

The final result is
$$
{
\mathrm{QFT}_4\left(\frac{|01\rangle+|11\rangle}{\sqrt2}\right)
=\frac{|00\rangle-|10\rangle}{\sqrt2}.
}
$$

## Implementation in Python
There are several ways to implement QPE in Python, and each of them can be applied to different situations. Here is a simple example using NumPy to create a QFT function that computes the same examples as above according to the formula.

```python
import numpy as np
# We import the NumPy library to conveniently work with arrays and complex numbers.

def qft(x):
    x = np.array(x, dtype=complex)
    # We convert the input x into a NumPy array with complex type.
    N = len(x)
    y = np.zeros(N, dtype=complex)
    # We create an empty result array y with N elements.
    # Initially, all of them are 0+0j, i.e. complex zeros.
    for k in range(N):
        # The loop goes through all output indices k = 0..N-1.
        # We will calculate each y_k separately.
        s = 0
        # s is the sum.
        for j in range(N):
            # The inner loop goes through all input indices j = 0..N-1.
            s += np.exp(2j * np.pi * j * k / N) * x[j]
        y[k] = s / np.sqrt(N)
    return y
    # We return the full result vector y = [y_0, y_1, ..., y_{N-1}]

# Example |11>
x_11 = [0, 0, 0, 1]
y_11 = qft(x_11)
print("Input |11> amplitudes:", x_11)
print("QFT(|11>) amplitudes:", np.round(y_11, 3)) 


# Example (|01> + |11>)/sqrt(2)
x_super = [0, 1/np.sqrt(2), 0, 1/np.sqrt(2)]
y_super = qft(x_super)
print("\nInput (|01>+|11>)/sqrt2 amplitudes:", x_super)
print("QFT(result) amplitudes:", np.round(y_super, 3))
```
